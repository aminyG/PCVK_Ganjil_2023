###Creating 2D Geometry image using OpenCV

```python
black_img = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(black_img)
```

```python
#perhatikan koordinat titik2 pt1 dan pt2
cv.rectangle(black_img,pt1=(384,0), pt2=(510,150), color=(0,255,0), thickness=10)
plt.imshow(black_img)
```

```python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300), color=(0,0,255), thickness=15)
plt.imshow(black_img)
```

```python
cv.circle(black_img,center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(black_img)
```

```python
cv.circle(black_img,center=(400,400),radius=30,color=(255,0,0),thickness=60)
cv.line(black_img, pt1=(0,0), pt2=(512,512), color=(255,255,255), thickness=5)
plt.imshow(black_img)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello', org=(10,500), fontFace=font, fontScale=4, color=(255,255,0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_img)
```

```python
black_img2=np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(black_img2)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices
```

```python
pts = vertices.reshape((-1,1,2)) #nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel
pts
```

```python
cv.polylines(black_img2,[pts],isClosed=True, color=(255,0,0), thickness=5)
plt.imshow(black_img2)
```

### Question
1. What is the difference between the images displayed without and with matplotlib?

The key distinction between image display in OpenCV and Matplotlib is their default color representation. OpenCV uses the BGR (Blue, Green, Red) format, whereas Matplotlib and most other image processing libraries use RGB (Red, Green, Blue).

2. What is the difference and effect of creating a black image between int16 and int32 data types?

The difference between creating a black image with int16 and int32 data types mainly lies in the range of values and memory usage. Int16 is memory-efficient for black images within its range, while int32 may be preferred for wider value ranges and extensive computations to avoid overflow and data loss.

The "from google.colab.patches import cv2_imshow" code imports the cv2_imshow function from Google Colab's patches module, enhancing Jupyter notebook functionality in the Colab environment by facilitating image display directly in output cells.
