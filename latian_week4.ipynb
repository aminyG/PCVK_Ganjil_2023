```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 20x20 yellow image
# width, height = 20, 20
yellow_color = (1, 1, 0)  # RGB color code for yellow
yellow_image = np.ones((20, 20, 3)) * yellow_color

# Display the yellow image
plt.imshow(yellow_image)
plt.axis('on')
plt.show()
```
